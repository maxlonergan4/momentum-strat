```python
import yfinance as yf
import pandas as pd

# Download Apple (AAPL) stock data
data = yf.download('AAPL', start='2022-01-01', end='2023-01-01')

# Show the first few rows of data
data.head()
```

```python
# Calculate 20-day and 50-day moving averages of closing prices
data['Short_MA'] = data['Close'].rolling(window=20).mean()
data['Long_MA'] = data['Close'].rolling(window=50).mean()

# Show the updated data with moving averages
data[['Close', 'Short_MA', 'Long_MA']].head(60)
```

```python
# Initialize the 'Signal' column: 1 means "buy", 0 means "sell/hold"
data['Signal'] = 0

# When short MA > long MA, signal = 1 (buy), else 0
data.loc[data['Short_MA'] > data['Long_MA'], 'Signal'] = 1

# Create 'Position' column to spot changes in signals (buy = 1, sell = -1)
data['Position'] = data['Signal'].diff()

# Show the last few rows to see the signals
data[['Close', 'Short_MA', 'Long_MA', 'Signal', 'Position']].tail(20)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))
plt.plot(data['Close'], label='Price', alpha=0.5)
plt.plot(data['Short_MA'], label='20-day MA')
plt.plot(data['Long_MA'], label='50-day MA')

# Plot buy signals (where Position = 1)
plt.plot(data[data['Position'] == 1].index, 
         data['Short_MA'][data['Position'] == 1], 
         '^', markersize=10, color='g', label='Buy Signal')

# Plot sell signals (where Position = -1)
plt.plot(data[data['Position'] == -1].index, 
         data['Short_MA'][data['Position'] == -1], 
         'v', markersize=10, color='r', label='Sell Signal')

plt.legend()
plt.title("Momentum Strategy on AAPL")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```
